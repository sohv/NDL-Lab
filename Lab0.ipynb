# Activation functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sigmoid function

```python
def sig(x):
    return 1/(1+np.exp(-x))
```

```python
sig(1.0)
```

```python
# plotting the graph
x = np.linspace(-10, 10, 100)
y=sig(x)
plt.plot(x,y)
plt.title('Sigmoid activation functions')
plt.xlim([-11,11])
plt.ylim([-0.1,1.1])
plt.grid(True)
plt.show()
```

## ReLU function

```python
def relu(x):
    return np.maximum(0,x)
```

```python
x = np.linspace(-5, 5, 100)
y=relu(x)
plt.plot(x,y)
plt.title('ReLU')
plt.grid(True)
plt.show()
```

## Leaky ReLU

```python
def leaky_relu(x):
    return np.maximum(0.01*x, x)
```

```python
x = np.linspace(-10, 10, 100)
y=leaky_relu(x)
plt.plot(x,y)
plt.title('Leaky ReLU')
plt.grid(True)
plt.show()
```

## Tanh function

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10, 10, 100)
y=tanh(x)
plt.plot(x,y)
plt.title('Tanh')
plt.grid(True)
plt.show()
```

## Softmax

```python
def softmax(x):
    exp_scores = np.exp(x)
    return exp_scores / np.sum(exp_scores)
```

```python
x = np.linspace(-10, 10, 100)
y = softmax(x)
plt.plot(x,y)
plt.title('Softmax')
plt.grid(True)
plt.show()
```

## Binary function

```python
def binary(x):
    return np.where(x>=0, 1, 0)
```

```python
x = np.linspace(-15, 10, 100)
y=binary(x)
plt.plot(x,y)
plt.title('Binary')
plt.grid(True)
plt.show()
```

## Linear function

```python
def linear(x):
    return x
```

```python
x = np.linspace(-5, 5, 100)
y=linear(x)
plt.plot(x,y)
plt.title('Linear')
plt.grid(True)
plt.show()
```
